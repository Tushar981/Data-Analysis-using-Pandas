```python
#.apply ()-> it willgoing to apply string at the end of number

def convert_Number_to_String(number):
    return str(number)+"Millions!"
    
columns=["Box_Office","Budget","Bond_Actor_Salary"]
for col in columns:
    bond[col]=bond[col].apply(convert_Number_to_String)
```

```python
bond.head(5)
```

```python
#Now i want to apply the things on the vertical way not the horrizintal way

def good_movie(row):
    actor=row[1]
    budget=row[4]
    
    if actor=="Pierce Brosnan":
        return "The Best"
    elif actor=="Rooger Moore" and budget>40:
        return "Enjoyable"
    else:
        return"I have no clue"

bond.apply(good_movie,axis="columns")
```
