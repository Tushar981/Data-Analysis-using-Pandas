# grouping by multiple column

```python
import pandas as pd
fortune=pd.read_csv("fortune1000.csv",index_col="Rank")
Sector=fortune.groupby(["Sector","Industry"])
fortune.head()
```

```python
Sector.size()
```

```python
Sector.sum()
```

```python
# we can also put different different aggregation rule on each and indivitual collumn 
Sector.agg({
    "Revenue":"sum",
    "Profits":"sum",
    "Employees":"mean"
})
```

```python
Sector.agg({
    "Revenue":["sum","mean"],
    "Profits":"sum",
    "Employees":"mean"
})
```

```python
Sector.agg({
    "Revenue":["size","sum","mean"],
    "Profits":["size","sum","mean"],
    "Employees":["size","sum","mean"]
})
```

```python
## Iterating Over Groups
fortune =pd.read_csv("fortune1000.csv",index_col="Rank")
sectors=fortune.groupby("Sector")
sectors.head()
```

```python
df=pd.DataFrame(columns=fortune.columns)
df
```

```python
import pandas as pd
#so whenever u are iterating on the groupby that time basically required two parameters 
#1 is group ->that represent group
#2 is dataframe -->where the actual dataframe is stored.
for sect,data in sectors:
    highest_revenue_company_in_group=data.nlargest(1,"Revenue")# 1 is represent 1 st value and second parameter is on which column
    df=df.append(highest_revenue_company_in_group)
```

```python
cities=fortune.groupby("Location")
df=pd.DataFrame(columns=fortune.columns)
df
```

```python
for cities,data in cities:
    highest_revenue_in_city=data.nlargest(1,"Revenue")
    df=df.append(highest_revenue_in_city)
```
