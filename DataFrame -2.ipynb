```python
#.Duplicate()->this will provide ypu the result in the first row it will provide i never seen that result but from 2 row if we are getting that value in that case it will going to consider as duplicate 
df.sort_values("First Name",inplace=True)
```

```python
df["First Name"].duplicated()
```

```python
#Keep=first ->first will be new and remaining will be duplicate
#Keep=last->last will be new and remaining(upper one) will be duplicate
#Keep=false->doesn't matter the value coming first time or last time if it is duplicate it will show the True 
df["First Name"].duplicated(keep="last")
```

```python
#.drop_duplicates -> it helps to remove the duplicate from the dataframe
len(df)
len(df.drop_duplicates())
#like that we can not delete cause every row data is identical so it is not possible so now we need to remove the data on column basis
```

```python
df.drop_duplicates(subset=["First Name"],keep="first")
df.drop_duplicates(subset=["First Name"],keep="last")
df.drop_duplicates(subset=["First Name"],keep=False)
```
