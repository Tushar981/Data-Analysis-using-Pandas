```python
import pandas as pd
from pandas_datareader import data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url="https://data.cityofnewyork.us/resource/25th-nujf.csv"
baby_data=pd.read_csv(url)
baby_data.head()
```

```python
graph_data=baby_data[["rnk","cnt"]]
```

```python
graph_data.plot()
```

```python
graph_data.plot(y="rnk")
graph_data["rnk"].plot()
```
