# Change index level name with the set_names()

```python
import pandas as pd
bigmac=pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.index.set_names(names=["Day","Location"],inplace=True)
```

```python
bigmac.head()
```

```python
#but if i wamt to change the value of Day ->Date and Location ->country agin in that case we need to set level in the set_names()
bigmac.index.set_names(names="Date",level=0)
bigmac.index.set_names(names="Date",level="Day") # so the meaning is we want to change from Day to Date
```

```python
bigmac.index.set_names(names="Location",level=1)
bigmac.index.set_names(names="Country",level="Location")
```

# Sort_index() on multiindex Dataframe


```python
import pandas as pd
bigmac=pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.sort_index()
bigmac.sort_index(ascending=True)
bigmac.sort_index(ascending=False)

#this will going to a make the changes on first column will chnage in ascending and second column will going to change in desceding order
bigmac.sort_index(ascending=[True,False])
```

```python
# instead of that we can also define the levels
bigmac.sort_index(level=0)
bigmac.sort_index(level="Date")
bigmac.sort_index(level=1)
bigmac.sort_index(level="Country")
```


# Swap level Method

```python
import pandas as pd
bigmac=pd.read_csv("bigmac.csv",parse_dates=["Date"],index_col=["Date","Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
# below all way will going to swap the index it doesnt matter u are writtering in which pattern 
# there is no inplace variable in function so if we want to fix that time time we need define into the bigmac="     "
bigmac.swaplevel()

bigmac.swaplevel("Date","Country")
bigmac.swaplevel("Country","Date")

bigmac.swaplevel(0,1)
bigmac.swaplevel(1,0)
```

# The Stack Method

```python
worldstats=pd.read_csv("worldstats.csv",index_col=["country","year"])
worldstats.head()
```

```python
#the stack method basically what it does population and GDP will going to shift in row .So overall idea is column header will be
# converted into row 
worldstats=worldstats.stack()
worldstats.head()
```

```python
type(worldstats)
# so this is basically a series so in that casse what we need to convert them in dataframe.
```

```python
frame=worldstats.to_frame()
frame.head()
```

# The .unstack method Part -1

```python
#.unstack() will going to reverse the stack()

worldstats.unstack()
```

```python
#so if i am going  to implement the unstack method again then it will going to transpose the index
# what it will do so in the above result we have two index so second index will going to convert into the header based on that 
#it will going to show the result.
worldstats.unstack().unstack()
```

```python
# use of level parameter
worldstats.unstack(level=["year","country"])
```

# The pivot_table()

```python
# pivot_table fuction is more powerful as compare to pivot function.

import pandas as pd
foods=pd.read_csv("foods.csv")
foods.head()
```

```python
foods.pivot_table(index="Gender",values="Spend",columns="City",aggfunc="mean")
foods.pivot_table(index="Gender",values="Spend",columns="City",aggfunc="sum")
```

```python
foods.pivot_table(index=["Gender","Item"],values="Spend",columns="City",aggfunc="sum")
```

```python
# there is two way to write the below statements we can call using the CSV table name variable name 
foods.pivot_table(index=["Gender","Item"],values="Spend",columns="City",aggfunc="min")
```

```python
# and also we can call using pandas variable name also
pd.pivot_table(data=foods,index=["Gender","Item"],values="Spend",columns="City",aggfunc="min")
```

# The .pd_melt()

```python
#this helps us to convert data into tabular 
import pandas as pd
sales=pd.read_csv("quarters.csv")
sales.head()
```

```python
#parameter defination ->
#frame(1 st parameter) means table where we store the table
#id_vars->the table is to be perserved it means that the column we need to keep it 
#var_name->it means whatever column name is genrated we can changes it with new number "Quarter"
#value_name->"Change the value column name "

pd.melt(sales,id_vars="Salesman")
```

```python
pd.melt(sales,id_vars="Salesman",var_name="Quarter",value_name="Measure Value")
```
