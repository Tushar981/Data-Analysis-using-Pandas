# Export Excel File

```python
url="https://data.cityofnewyork.us/resource/25th-nujf.csv"
baby_name=pd.read_csv(url)
baby_name.head()
```

```python
male=baby_name[baby_name["gndr"]=="MALE"]
female=baby_name[baby_name["gndr"]=="FEMALE"]
```

```python
female
```

```python
male
```

```python
excel_file=pd.ExcelWriter("Baby_name.xlsx")
```

```python
female.to_excel(excel_file,sheet_name="Male",index=False)
male.to_excel(excel_file,sheet_name="Female",index=False)
```

```python
excel_file.save()
```
