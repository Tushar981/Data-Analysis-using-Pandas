```python
## Next Section 
# we are using dropna how ="all" cause in the end one cell row containing complete null values
import pandas as pd
chicago=pd.read_csv("chicago.csv").dropna(how="all")
chicago.tail(3)
```

```python
chicago["Department"]=chicago["Department"].str.replace("MGMNT","MANAGEMENT")
chicago["Employee Annual Salary"]=chicago["Employee Annual Salary"].str.replace("$","").astype(float)
```

```python
chicago["Employee Annual Salary"].sum()
chicago["Employee Annual Salary"].mean()
```

# Filtering with string method

```python
#starts with 
#end with
#contain

chicago[chicago["Position Title"].str.lower().str.contains("water")]
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
chicago[chicago["Position Title"].str.lower().str.endswith("ist")]


```

```python
# strip()->remove the whitespace from the left and right 
#lstrip()
#rstrip()
chicago["Name"]=chicago["Name"].str.strip()
chicago["Position Title"]=chicago["Position Title"].str.strip()
```

```python
# .split()->method helps us to seprate the name based on the strip condition
import pandas as pd
chicago=pd.read_csv("chicago.csv").dropna(how ="all")
chicago.head(3)
chicago["Name"].str.split(",").str.get(0).value_counts()
```
